### High Boost Filter

The high boost filter can "repair" or to get better an image from a "blur" version of that image.

$$ Better = Original - BlurVersion $$

To do this, the kernel is the shown below:

$$ K = \begin{bmatrix} -1 & -1 & -1 \\ -1 & A+8 & -1 \\ -1 & -1 & -1 \end{bmatrix} $$

Being A > 1 or A = 1 (Laplacian case)

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

enCasa = cv.imread('../data/Images/cameraman.tif', cv.IMREAD_GRAYSCALE)
plt.imshow(enCasa, cmap = 'gray')
cv.imshow(winname='bad quality photo', mat=enCasa)
cv.waitKey(0)
```

```python
As = [1.0, 1.4, 1.8, 2.2]
K = [np.array([[-1.0, -1.0, -1.0], [-1.0, A+8.0, -1.0], [-1.0, -1.0, -1.0]]) for A in As]

for k in K:
    filtered = cv.filter2D(enCasa, -1, k)
    name_k = 'A = ' + str(k)
    #cv.imshow(winname=name_k, mat = filtered)
    #cv.waitKey(0)
    plt.figure()
    plt.imshow(filtered, cmap = 'gray')
```
